```python
import pandas as pd
from faker import Faker
import random
import streamlit as st
```

```python
faker = Faker()
n = 10
```

```python
Customers = []

for _ in range(n):
  Customers.append({
      'customer_id': faker.uuid4(),
      'name': faker.name(),
      'email': faker.email(),
      'phone_number': faker.phone_number(),
      'location': faker.city(),
      'signup_date': faker.date_this_decade(),
      'is_premium': faker.boolean(),
      'preferred_cuisine': random.choice(['Indian', 'Chinese', 'Mexican', 'Italian', 'Arabian']),
      'total_orders': random.randint(1, 10),
      'average_rating': round(float(random.randint(1, 5)),2)
  })

df1 = pd.DataFrame(Customers)
df1
```

```python
Restaurants = []

for _ in range(n):
  Restaurants.append({
      'restaurant_id': faker.uuid4(),
      'name': faker.company(),
      'cuisine_type': random.choice(['Indian', 'Chinese', 'Mexican', 'Italian', 'Arabian']),
      'location': faker.city(),
      'owner_name': faker.name(),
      'average_delivery_time': random.randint(10, 60),
      'contact_number': faker.phone_number(),
      'rating': round(float(random.randint(1, 5)),2),
      'total_orders': random.randint(1,50),
      'is_active': faker.boolean()
  })

df2 = pd.DataFrame(Restaurants)
df2
```

```python
Orders = []

for _ in range(n):
  Orders.append({
      'order_id': faker.uuid4(),
      'customer_id': random.choice(Customers)['customer_id'],
      'restaurant_id': random.choice(Restaurants)['restaurant_id'],
      'order_date': faker.date_time_this_month(),
      'delivery_time': faker.date_time_this_month(),
      'status': random.choice(['Pending','Delivered','Cancelled']),
      'total_amount': random.randint(100, 1000),
      'payment_mode': random.choice(['Credit Card','Cash','UPI']),
      'discount_applied': random.choice([10,20,30,40,50,60,70,80]),
      'feedback_rating': round(float(random.randint(1, 5)),2)
  })

df3 = pd.DataFrame(Orders)
df3
```

```python
Deliveries = []

for _ in range(n):
  Deliveries.append({
      'delivery_id': faker.uuid4(),
      'order_id': random.choice(Orders)['order_id'],
      'delivery_status': random.choice(['On the Way', 'Delivered', 'Cancelled']),
      'distance': random.randint(1,20),
      'delivery_time': random.randint(10,60),
      'estimated_time': random.randint(10,60),
      'delivery_fee': random.randint(10,50),
      'vehicle_type': random.choice(['Bike','Scooter'])
  })

df4 = pd.DataFrame(Deliveries)
df4
```

```python
df1.to_csv('Customers.csv', index=False)
df2.to_csv('Restaurants.csv', index=False)
df3.to_csv('Orders.csv', index=False)
df4.to_csv('Deliveries.csv', index=False)
```
